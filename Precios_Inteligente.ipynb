Para reemplazar los simbolos utilizamos la funcion apply que nos ayuda a aplicar cosas dentro del df.. en este codigo estamos reemplazando el signo $ por nada. x se utiliza por convencion
["nombre de columna"].apply(Lambda x: x.replace("$","").

```python
datos_hosting["cuota_limpieza"] = datos_hosting["cuota_limpieza"].apply(lambda x: x.replace("$","").replace(",","").strip())
datos_hosting["cuota_deposito"] = datos_hosting["cuota_deposito"].apply(lambda x: x.replace("$","").replace(",","").strip())
```

```python
datos_hosting["cuota_limpieza"]= datos_hosting["cuota_limpieza"].astype(np.float64)
datos_hosting["cuota_deposito"]= datos_hosting["cuota_deposito"].astype(np.float64)
```
